```python
from os import path, getcwd, listdir
from hundir_la_flota.hundir_la_flota_con_clases import *
import neat
import random
import pickle
```

```python
def eval_genomes(genomes, config):
    # View at the end
    display = False
    displayAll = False
    
    nets = []
    ge = []
    almirantes = []
    
    for genome_id, genome in genomes:
        genome.fitness = 0
        net = neat.nn.FeedForwardNetwork.create(genome, config)
        nets.append(net)
        almirantes.append(Almirante("IA", True))
        ge.append(genome)
        
    for indice, almirante in enumerate(almirantes):
        # Variables de juego
        turno = 1
        score = 0
        numero_de_disparos = 0
        posicions_conocidas = [(i, j) for i in range(10) for j in range(10)]
        end = True
        
        while (end):
            # Elegir una coordenada para disparar
            tablero_enemigo_numpy = almirante.tablero_enemigo_para_IA()
            output = nets[almirantes.index(almirante)].activate(tablero_enemigo_numpy)
            fila, columna = almirante.salida_de_red_a_indice(np.argmax(output))

            if (fila, columna) in posicions_conocidas:
                resultado_del_disparo = almirante.recibir_disparo(fila, columna)
                posicions_conocidas.remove((fila, columna))

            else:
                ge[indice].fitness -= 1
                if len(posicions_conocidas) > 0:
                    fila, columna = posicions_conocidas[-1]
                    posicions_conocidas.pop()
                    resultado_del_disparo = almirante.recibir_disparo(fila, columna)
                    
            if resultado_del_disparo == "A":
                ge[indice].fitness -= 1
                score -= 1
            elif resultado_del_disparo == "T":
                ge[indice].fitness += 10
                score += 10
            elif resultado_del_disparo == "H":
                ge[indice].fitness += 10
                score += 10
            numero_de_disparos += 1

            # Actualizar mapa con (Tocado, Hundido y Agua)
            almirante.actualizar_tras_disparo(resultado_del_disparo, fila, columna)

            # Combrobar si hay ganador, 20 gana
            if almirante.soy_ganador():
                ge[indice].fitness += 20
                score += 5
                #score += 5
                #almirante.imprimir_tableros()
                #print(f"Gana {almirante.nombre}")
                #print(f"{ge[indice].fitness}")
                #print(f"Score:{score}")
                #print(f"Disparos:{numero_de_disparos}")
                #print()
                #print()
                end = False
```

```python
def run(config_file):
    """
    runs the NEAT algorithm to train a neural network to play flappy bird.
    :param config_file: location of config file
    :return: None
    """
    config = neat.config.Config(neat.DefaultGenome, neat.DefaultReproduction,
                         neat.DefaultSpeciesSet, neat.DefaultStagnation,
                         config_file)

    # Create the population, which is the top-level object for a NEAT run.
    p = neat.Population(config)

    # Add a stdout reporter to show progress in the terminal.
    p.add_reporter(neat.StdOutReporter(False))
    stats = neat.StatisticsReporter()
    p.add_reporter(stats)
    #p.add_reporter(neat.Checkpointer(5))
    gensToFile = 200
    # Run for up to 50 generations.
    winner = p.run(eval_genomes, gensToFile)
    with open(f"red_mamadisima_gens{gensToFile}_10layers.pkl", "wb") as f:
        pickle.dump(winner, f)
        
    
    # show final stats
    print('\nBest genome:\n{!s}'.format(winner))
```

```python
if __name__ == '__main__':
    # Determine path to configuration file. This path manipulation is
    # here so that the script will run successfully regardless of the
    # current working directory.
    config_path = path.join(getcwd(), 'hundir_la_flota/config-feedforward.txt')
    run(config_path)
```

```python
def multiples_Almirantes(display=False, displayAll=False):
    almirantes = [Almirante(f"IA {_}") for _ in range(0, 10)]
    enemigo = Almirante("Bot")

    for almirante in almirantes:
        # Variables de juego
        turno = 1
        score = 0
        numero_de_disparos = 0
        posicions_conocidas = [(i, j) for i in range(10) for j in range(10)]
        random.shuffle(posicions_conocidas)
        end = True
        
        while (end):
            # Elegir una coordenada para disparar
            fila, columna = posicions_conocidas[-1]
            posicions_conocidas.pop()
            resultado_del_disparo = enemigo.recibir_disparo(fila, columna)
            if resultado_del_disparo == "A":
                score -= 1
            elif resultado_del_disparo == "T":
                score += 4
            elif resultado_del_disparo == "H":
                score += 4
            numero_de_disparos += 1

            # Actualizar mapa con (Tocado, Hundido y Agua)
            if turno == 1:
                almirante.actualizar_tras_disparo(resultado_del_disparo, fila, columna)

            if display and displayAll:
                # Impresi√≥n de los tableros
                if turno == 1:
                    almirante.imprimir_tableros()

            # Combrobar si hay ganador, 20 gana
            if turno == 1:
                if almirante.soy_ganador() and display:
                    score += 5
                    #score += 5
                    #almirante.imprimir_tableros()
                    #print(f"Gana {almirante.nombre}")
                    print(f"Score:{score}")
                    print(f"Disparos:{numero_de_disparos}")
                    print()
                    #print()
                    end = False

            # Velocidad del juego
            #sleep(0.1)
multiples_Almirantes(True)
```
